```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
```

```python
from sklearn.cluster import KMeans
```

```python
x,y = make_moons(n_samples = 200, noise = 0.10)

```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(x)
```

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(x[:,0],x[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],color='red')
plt.show()
```

# DBSCAN
    - Density based Spatial Clustering of Application and Noise
    - Based on the idea that cluster is high density area surrounded by low density area
    - Starts by exploring a small area, if density is good enough it is considerd as a part of cluster and explores neighbours and so on, so it groups regions based on density.

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2,min_samples=5)
```

```python
dbs.fit(x)
```

```python
ypred = dbs.fit_predict(x)
```

```python
plt.scatter(x[: , 0] , x[:,1],c=ypred)
plt.show()
```
